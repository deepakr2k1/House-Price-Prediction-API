```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

### Read Data

```python
df = pd.read_csv('train-data.csv')
df.head()
```

```python
df.info()
```

```python
df.drop(['Unnamed: 0','New_Price'],axis=1,inplace = True)
```

```python
df['Mileage'] = df['Mileage'].str.split(pat=" ").str[0]
df['Engine'] = df['Engine'].str.split(pat=" ").str[0]
df['Power'] = df['Power'].str.split(pat=" ").str[0]
```

```python
df['Mileage'] = pd.to_numeric(df['Mileage'],errors='coerce')
df['Engine'] = pd.to_numeric(df['Engine'],errors='coerce')
df['Power'] = pd.to_numeric(df['Power'],errors='coerce')
```

```python
df['Mileage'].fillna(df['Mileage'].mean(),inplace = True)
df['Engine'].fillna(df['Engine'].mean(),inplace = True)
df['Power'].fillna(df['Power'].mean(),inplace = True)
df['Seats'].fillna(np.floor(df['Seats'].mean()),inplace = True)
```

```python
df['Year']=2020-df['Year']
```

```python
dummies0 = pd.get_dummies(df['Name'])
dummies1 = pd.get_dummies(df['Location'])
dummies2 = pd.get_dummies(df['Fuel_Type'])
dummies3 = pd.get_dummies(df['Transmission'])
dummies4 = pd.get_dummies(df['Owner_Type'])
```

```python
df = pd.concat([df,dummies0,dummies1,dummies2,dummies3,dummies4],axis=1)
```

```python
df = df.drop(['Name','Volvo XC90 2007-2015 D5 AWD','Location','Pune','Fuel_Type','Petrol','Transmission','Manual','Owner_Type','Fourth & Above'],axis=1)
```

```python
df
```

```python
x = df.drop(['Price'],axis=1)
y = df['Price']
x,y
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(x,y)
```

```python
test = pd.read_csv('test-data.csv')
test
```

```python
test = test.drop(['Unnamed: 0','New_Price'],axis=1)
```

```python
test['Mileage'] = test['Mileage'].str.split(pat=" ").str[0]
test['Engine'] = test['Engine'].str.split(pat=" ").str[0]
test['Power'] = test['Power'].str.split(pat=" ").str[0]
test['Mileage'] = pd.to_numeric(test['Mileage'],errors='coerce')
test['Engine'] = pd.to_numeric(test['Engine'],errors='coerce')
test['Power'] = pd.to_numeric(test['Power'],errors='coerce')
test['Mileage'].fillna(test['Mileage'].mean(),inplace = True)
test['Engine'].fillna(test['Engine'].mean(),inplace = True)
test['Power'].fillna(test['Power'].mean(),inplace = True)
test['Seats'].fillna(np.floor(test['Seats'].mean()),inplace = True)
```

```python
test['Year']=2020-test['Year']
test
```

```python
test_x = pd.DataFrame(columns = x.columns)
test_x['Year'] = test['Year']
test_x['Kilometers_Driven'] = test['Kilometers_Driven']
test_x['Mileage'] = test['Mileage']
test_x['Engine'] = test['Engine']
test_x['Power'] = test['Power']
test_x['Seats'] = test['Seats']
test_x
```

```python
for i in test_x.index:
    name = test['Name'].loc[i]
    if name in test_x.columns:
        test_x.loc[i,name]=1
        
    location = test['Location'].loc[i]
    if location in test_x.columns:
        test_x.loc[i,location]=1
    
    fuel_type = test['Fuel_Type'].loc[i]
    if fuel_type in test_x.columns:
        test_x.loc[i,fuel_type]=1
    
    transmission = test['Transmission'].loc[i]
    if transmission in test_x.columns:
        test_x.loc[i,transmission]=1
    
    owner_Type = test['Owner_Type'].loc[i]
    if owner_Type in test_x.columns:
        test_x.loc[i,owner_Type]=1    
```

```python
test_x.fillna(0,inplace=True)
```

```python
predicted_y = model.predict(test_x)
predicted_y
```
